```python
import pandas as pd

# Load the datasets
labor_force_df = pd.read_csv('Total labor force.csv')
unemployment_rate_df = pd.read_csv('Unemployment rate.csv')

# Define the list of 38 OECD countries
oecd_countries = [
    'Canada', 'Mexico', 'United States', 'Austria', 'Belgium', 'France', 'Germany', 'Ireland', 'Luxembourg',
    'Netherlands', 'Switzerland', 'United Kingdom', 'Denmark', 'Estonia', 'Finland', 'Iceland', 'Latvia',
    'Lithuania', 'Norway', 'Sweden', 'Greece', 'Italy', 'Portugal', 'Spain', 'Turkey', 'Czech Republic',
    'Hungary', 'Poland', 'Slovak Republic', 'Slovenia', 'Australia', 'Japan', 'South Korea', 'New Zealand',
    'Chile', 'Colombia', 'Costa Rica'
]

# Filter data for the specified OECD countries
labor_force_oecd = labor_force_df[labor_force_df["Country Name"].isin(oecd_countries)]
unemployment_rate_oecd = unemployment_rate_df[unemployment_rate_df["Country Name"].isin(oecd_countries)]

# Replace missing values with "No_Data"
labor_force_oecd.fillna("No_Data", inplace=True)
unemployment_rate_oecd.fillna("No_Data", inplace=True)

# Select only relevant years (2010 onwards)
columns_to_keep = ["Country Name", "Country Code", "Indicator Name", "Indicator Code"] + [str(year) for year in range(2010, 2024)]
labor_force_oecd = labor_force_oecd[columns_to_keep]
unemployment_rate_oecd = unemployment_rate_oecd[columns_to_keep]

# Save the cleaned datasets to new CSV files
labor_force_oecd.to_csv("Cleaned_Labor_Force_OECD.csv", index=False)
unemployment_rate_oecd.to_csv("Cleaned_Unemployment_Rate_OECD.csv", index=False)

print("Data cleaning complete. Cleaned files saved.")
x
```
